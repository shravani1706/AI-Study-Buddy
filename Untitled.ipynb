```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define nodes for each module in the system
nodes = [
    "User Input",
    "Streamlit Web Interface",
    "PDF Extraction & Preprocessing Module",
    "NLP Processing (spaCy)",
    "Feature Extraction (TF-IDF & Cosine Similarity)",
    "Generative AI Module (Gemini-1.5 Pro)",
    "Ranking Engine",
    "Results Display"
]

# Add nodes to the graph
G.add_nodes_from(nodes)

# Define edges (connections) between nodes
edges = [
    ("User Input", "Streamlit Web Interface"),
    ("Streamlit Web Interface", "PDF Extraction & Preprocessing Module"),
    ("PDF Extraction & Preprocessing Module", "NLP Processing (spaCy)"),
    ("NLP Processing (spaCy)", "Feature Extraction (TF-IDF & Cosine Similarity)"),
    ("Streamlit Web Interface", "Generative AI Module (Gemini-1.5 Pro)"),
    ("Feature Extraction (TF-IDF & Cosine Similarity)", "Ranking Engine"),
    ("Generative AI Module (Gemini-1.5 Pro)", "Ranking Engine"),
    ("Ranking Engine", "Results Display")
]

G.add_edges_from(edges)

# Generate positions for the nodes using a shell layout for a hierarchical appearance
pos = nx.shell_layout(G)

# Draw the nodes, edges, and labels
plt.figure(figsize=(12, 8))
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='lightblue')
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')

plt.title("Flowchart: AI-powered Resume Screening System", fontsize=16)
plt.axis('off')
plt.tight_layout()

# Save the flowchart as an image file
plt.savefig("flowchart_networkx.png", format="PNG")
plt.show()
```
